```python
import json
import pandas as pd
from pathlib import Path
```

```python
df = pd.read_parquet('/nas_train/app.e0016372/datasets/medmcqa/data/train-00000-of-00001.parquet')
df = df.sample(10000, random_state=42)
result = []
for i in df.index:
    instruction = df.loc[i, 'question']
    input = df.loc[i, 'choices'].tolist()
    output = input[df.loc[i, 'answer_index']]
    result.append({
        'instruction': instruction,
        'input': f'Choices: {input}',
        'output': output
    })
df = pd.DataFrame(result)
df.to_json('med_mcqa_train.jsonl', orient='records', force_ascii=False, lines=True)
df
```

```python
df = pd.read_parquet('/nas_train/app.e0016372/datasets/medmcqa/data/validation-00000-of-00001.parquet')
df = df.sample(1000, random_state=42)
result = []
for i in df.index:
    question = df.loc[i, 'question']
    choices = df.loc[i, 'choices']
    answer = df.loc[i, 'answer']
    result.append({
        'question': question,
        'A': choices[0],
        'B': choices[1],
        'C': choices[2],
        'D': choices[3],
        'answer': answer
    })
df = pd.DataFrame(result)
df.to_json('med_mcqa_val.jsonl', orient='records', force_ascii=False, lines=True)
df
```
